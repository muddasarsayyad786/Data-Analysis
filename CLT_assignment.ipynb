5. For billionaires dataset and titanic age column, first run shapiro test and qq-plot. Then try CLT on both and see if now we see normal distribution for their sample means.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
from scipy.stats import shapiro
```

```python
df1 = pd.read_csv('DataSets/titanic.csv')
df1.head()
```

```python
Shapiro_test = shapiro(df1['age'].fillna(df1['age'].mean()))
print("p-vaue of Shapiro_test:", Shapiro_test.pvalue)

alph = 0.05 #Common significance level
if Shapiro_test.pvalue > alph:
    print("Data is normally distributed")
else:
    print("Data is not normally distributed")
```

```python
from scipy import stats


stats.probplot(df1['age'], dist ="norm", plot = plt)
plt.title("QQ plot of Age")
plt.xlabel("Theoretical Quantiles(Normal Distribution)")
plt.ylabel("Sample Quantiles(Age)")
plt.grid()
plt.show()
```

```python
#apply Central limit theoram

sample_means = []
sample_count = 30
count = 25

for _ in range (count):
    sample_mean = df1['age'].fillna(df1['age'].mean()).sample(n=sample_count, replace = True)
    sample_means.append(sample_mean.mean())

print(sample_means)
```

```python
Shapiro_test = shapiro(sample_means)
print("p-vaue of Shapiro_test:", Shapiro_test.pvalue)

alph = 0.05 #Common significance level
if Shapiro_test.pvalue > alph:
    print("Data is normally distributed")
else:
    print("Data is not normally distributed")
```

```python
stats.probplot(sample_means, dist ="norm", plot = plt)
plt.title("QQ plot of Age")
plt.xlabel("Theoretical Quantiles(Normal Distribution)")
plt.ylabel("Sample Quantiles(Age)")
plt.grid()
plt.show()
```
