```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
df = pd.read_csv('DataSets/diabetes.csv')
df.head()
```

```python
df['z_score_glucose'] = zscore(df['Glucose'])
df.head()
```

```python
# outlier analysis using zscore
threeshould = 3
outliers_count = len(df[df['z_score_glucose'].abs() > threeshould])
print(outliers_count)
```

```python
#box plot
plt.Figure(figsize=(12,8))
sns.boxplot(y=df['Glucose'], color='lightblue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
print(df[df['z_score_glucose'].abs() > threeshould])
```

```python
df1 = pd.read_csv('DataSets/billionaires.csv')
print(df1.columns)
df1.head()
```

```python
df1['final_worth_zscore'] = zscore(df1['finalWorth'])
df1.head()
```

```python
# outlier analysis for billionaires final worth z-score
threshold = 3
outliers_count = len(df1[df1['final_worth_zscore'].abs() > threshold])
print("outliers_count:",outliers_count)
```

```python
plt.Figure(figsize=(12,8))
sns.boxplot(y=df1['finalWorth'], color='lightblue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
