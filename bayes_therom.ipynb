```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/titanic.csv')
total_length = len(df)

survived_count = len(df[df['survived'] == 1])
female_count = len(df[df['gender'] == 'female'])
female_survived_count = len(df[(df['survived'] == 1) & (df['gender'] == 'female')])

p_survived = survived_count/total_length
p_female = female_count/total_length
p_female_survived = female_survived_count/survived_count

#apply Bayes theorm
p_survived_given_female = (p_female_survived * p_survived) /p_female
print(p_survived_given_female)

```
