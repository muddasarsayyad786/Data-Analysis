```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("DataSets/Mall_Customers.csv")
df.head()
```

```python
#One Sample T-test
age_data = df['Age']
age_data = age_data.sample(n=30, replace=False)

pop_avg_age = 40 #H0: Mall customers are of avg age is 40

t_statists, p_value = stats.ttest_1samp(age_data, pop_avg_age)
print("P-value: ", p_value)

alpha = 0.05
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
#Two Sample T-test
data = pd.read_csv('DataSets/Mall_Customers.csv')
data.head()
```

```python
male_spending = data[data['Genre'] == 'Male' ]['Spending Score (1-100)'].sample(n=30,replace=False)
female_spending = data[data['Genre'] == 'Male' ]['Spending Score (1-100)'].sample(n=30, replace=False)

t_statists, p_value = stats.ttest_ind(male_spending, female_spending)
alpha = 0.05

print("P-value: ",p_value)

if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
#Paired Sample T_test

#paired t-test: A company has launched some offers
#H0: Pre and post usage is same (i.e. offers is not effective)

from scipy.stats import ttest_rel
import pandas as pd

df = pd.read_csv('DataSets/cust_seg.csv')
df.head()

usage_data = df[["pre_usage","Post_usage_1month"]].sample(n=30)

tstat,p_value = ttest_rel(usage_data['pre_usage'], usage_data['Post_usage_1month'])

alpha = 0.05

print("P-value: ",p_value)

if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
#H0: parantal education not effects on childs math marks
df = pd.read_csv('DataSets/StudentsPerformance.csv')
df.head()
```

```python
print(df['parental level of education'].unique())
```

```python
bachelor_degree = df[df['parental level of education'] == "bachelor's degree"]['math score']
some_college = df[df['parental level of education'] == "some college"]['math score']
master_degree = df[df['parental level of education'] == "master's degree"]['math score']
associate_degree = df[df['parental level of education'] == "associate's degree"]['math score']
high_school = df[df['parental level of education'] == "high school"]['math score']
some_high_school = df[df['parental level of education'] == "some high school"]['math score']

fstat, p_value = stats.f_oneway(bachelor_degree, some_college, master_degree, associate_degree, high_school, some_high_school)
print("P-value: ",p_value)

alpha = 0.05
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
#Chi-Square Test
from scipy.stats import chi2_contingency

df = pd.read_csv('DataSets/titanic.csv')
df.head()

#create contingency table
contingency_table = pd.crosstab(df['pclass'], df['survived'])
print(contingency_table)

#perfrom chi-square test
chi2_stats,p_value,dof,expected = chi2_contingency(contingency_table)

print("Chi-square statistic: ",chi2_stats)
print("P-Value: ",p_value)
print("Degree of freedom: ", dof)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```
