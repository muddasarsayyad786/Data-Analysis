```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("shopping.csv")
print(df.columns)
```

```python
#Ensure numeric coloms have correct data type
numeric_cols = ['CustomerID', 'CustomerID', 'Transaction_ID', 'Quantity', 'Avg_Price' 'Delivery_Charges', 'GST', 'Offline_Spend', 'Online_Spend', 'Month', 'Discount_pct']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col]) # error = coerce means conver to Nan if error
```

```python
#1) filter row by single value colom
# male_customes = df[df['Gender']== 'M']
# print(male_customes)

#altr method
male_customes = df.query("Gender == 'M'")
print(male_customes)
```

```python
#2) filter row by multiple colum condition
# male_chicago = df[(df['Gender']=='M') & (df['Location']=='Chicago')]
# print(male_chicago)

#alter
male_chicago = df.query("Gender = 'M' and Location = 'Chicago' ")
```

```python
#3 filter row using is in
```

```python
#4 Filer rows by numeric condition
# multiple_qty = df[df['Quantity']>1]
# print(multiple_qty)

multiple_qty = df.query("Quantity > 1")
print(multiple_qty)
```

```python
#5) Filter row using between 
# tenure_range=df[df['Tenure_Months'].between(6,12)]
# print(tenure_range)

# tenure_range=df[df['Tenure_Months'].between(6,12, inclusive = "left")] #inclusive = "left"/right/neither

tenure_range = df.query("6<= Tenure_Months <= 12")
print(tenure_range)
```

```python
#6) filter rows with string methods (contains) - safe for NaNs
camera_products = df[df['Product_Description'].str.contains('Camera', case = False, na = False)]
print(camera_products)

pd.set_option('display.max_columns',None) # gives all colums but so expensive
pd.reset_option('display.max_columns',None)# reset above setting

```

```python
#7 Filter rows based on null or non-null values
# with_coupon = df[df['Coupon_Code'].notnull()]
# print(with_coupon)

with_coupon= df.query("Coupon_Code == Coupon_Code")
print(with_coupon)
```

```python
#8) Filter rows by date range
df['Transaction_Date']= pd.to_datetime(df['Transaction_Date'])
# jan_2019 = df[(df['Transaction_Date']>='2019-01-01') & (df['Transaction_Date']<='2019-01-31')]
# print(jan_2019)

jan_2019 = df.query("Transaction_Date > = '2019-01-01' and Transaction_Date <= '2019-01-31'  ")
print(jan_2019)
```

```python
#9) extract specific colums after flitering

male_products= df[df['Gender'] == 'M'] [['CustomerID', 'Product_Description', 'Quantity']]
print(male_products)
```

```python
# #10) Filter top N values using nlargest
# top_online_spend = df.nlargest(5, 'Online_Spend')
# print(top_online_spend)

top_online_spend = df.sort_values('Online_Spend', ascending = False).head(5)
print(top_online_spend)
```

```python
# Filter columns with startwith / endwith - safe for NaNs
# electronic_products = df[df['Product_Description'].str.startswith('Nest', na = False)]
# print(electronic_products)

electronic_products = df[df['Product_Description'].str.endswith('USA', na = False)]
print(electronic_products)
```

```python
#12) Group-based filtering
customers_total_online = df.groupby('CustomerID')['Online_Spend'].sum()
# print(customers_total_online)
high_spender = customers_total_online[customers_total_online>5000]
print(high_spender)
```
