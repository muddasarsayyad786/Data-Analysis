```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataSets/weight-height.csv')
df.head()
```

```python
df_male = df[df['Gender']== 'Male']
df_female = df[df['Gender']== 'Female']
```

```python
corr_overall = df[['Height','Weight']].corr()
corr_male = df_male[['Height','Weight']].corr()
corr_female = df_female[['Height','Weight']].corr()

print("Overall Correlation:\n", corr_overall)
print("Male Correlation:\n", corr_male)
print("Female Correlation:\n", corr_female)
```

```python
cov_overall = df[['Height','Weight']].cov()
cov_male = df_male[['Height','Weight']].cov()
cov_female = df_female[['Height','Weight']].cov()

print("Overall Covariance:\n", cov_overall)
print("Male Covariance:\n", cov_male)
print("Female Covariance:\n", cov_female)
```
