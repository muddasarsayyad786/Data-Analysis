```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DataSets/tips.csv")
```

```python
#Histogram of total_bil-shows the distribution of single variable
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'],bins=10, color = 'skyblue',edgecolor='black')
plt.title('Distribution of Total Bill')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
#same plot using seaborn
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], bins = 10, kde = True, color="darkgreen") #kde=kernal Density (it shows a cntiue line over the histogram) 
plt.title('Distribution of Total Bill')
plt.show()
```

```python
#2) Bar plot of avarage tip by gender
avg_tip_gender = df.groupby('gender')['tip'].mean() # grouping automatically create index on grouped cloumn 
plt.figure(figsize=(6,4))
plt.bar(avg_tip_gender.index, avg_tip_gender.values, color=['pink','red']) #index = gender
plt.title('Avg tip by gender')
plt.ylabel('Avg Tip')
plt.show()
```

```python
#same with seaborn
avg_tip_gender = df.groupby('gender')['tip'].mean().reset_index() # grouping automatically create index on grouped cloumn 
plt.figure(figsize=(6,4))
sns.barplot(data = avg_tip_gender, x= 'gender', y= 'tip', palette=['pink','red']) #index = gender
plt.title('Avg tip by gender')
plt.ylabel('Avg Tip');
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(6,4))
plt.scatter(df['total_bill'], df['tip'], color='orange', edgecolor='black')
plt.title("Total bill vs Tip")
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
#scatter plot with sns
plt.figure(figsize=(6,4))
sns.regplot(x='total_bill', y='tip', color='orange', data = df)
plt.title("Total bill vs Tip with regression line")
# plt.xlabel('Total bill')
# plt.ylabel('Tip')
plt.show()
```
