```python
df = pd.read_csv('Datasets/income.csv')
df.head()
```

```python
#simple sampling method (select 10% of record randomly)
sample_df = df.sample(frac = 0.1, random_state=42) #frac 0.1 for getting 10% data and randon_state to keep random data fix
print("Sample data shape: ",sample_df.shape)
print("Population Data shape: ",df.shape)
df.head()
```

```python
#Systematic Ramdom Sampling Method
sample_df1 = df.iloc[::10,::] #df.iloc[row_indexer(start:end:step), column_indexer(start:end:step)]
print("Sample data shape: ",sample_df1.shape)
print("Population Data shape: ",df.shape)
df.head()
```

```python
white_count = len(df[df['race']=='White'])
print("Count of white race in population data: ", white_count)
Black_count = len(df[df['race']=='Black'])
print("Count of black race in population data: ", Black_count)

#sample data of white race
sample_white = df[df['race']=='White'].sample(frac=0.1, random_state=42)
print("Count of white   race in sample data: ", len(sample_white))

#sample data of black race
sample_black = df[df['race']=='Black'].sample(frac=0.1, random_state=42)
print("Count of black race in sample data: ", len(sample_black))

#sample data of Asian-Pac-Islander race
sample_asian = df[df['race']=='Asian-Pac-Islander'].sample(frac=0.1, random_state=42)
print("Count of Asian-Pac-Islander race in sample data: ", len(sample_asian))

#sample data of Amer-Indian-Eskimo race
sample_american = df[df['race']=='Amer-Indian-Eskimo'].sample(frac=0.1, random_state=42)
print("Count of Amer-Indian-Eskimo race in sample data: ", len(sample_american))

#sample data of Other race
sample_other = df[df['race']=='Other'].sample(frac=0.1, random_state=42)
print("Count of Other race in sample data: ", len(sample_other))

#concatenate sample data
sample_df2 = pd.concat([sample_white, sample_black, sample_asian, sample_american, sample_other], axis=0)
print("Shape of sample data after combining all: ", sample_df2.shape)
sample_df2.head()
```

```python
#cluster Smapling : as maratrial status
married_df = df['marital.status'].unique()
print("Unique marital status in population data: ", married_df)

#select 2 cluster randomly
selected_clusters = np.random.choice(married_df, size=3, replace=False)
print("Selected clusters: ", selected_clusters)
#select data of selected clusters
cluster_sample_df = df[df['marital.status'].isin(selected_clusters)]


print("Shape of sample data after cluster sampling: ", cluster_sample_df.shape)
cluster_sample_df.head()
```
