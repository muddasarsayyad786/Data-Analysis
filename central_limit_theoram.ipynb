```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
from scipy.stats import shapiro
```

```python
df = pd.read_csv('DataSets/diabetes.csv')
df.head()
```

```python
shario_test = shapiro(df['Glucose'])
print("Shapiro-Wik p-vale: ", shario_test.pvalue)

if shario_test.pvalue < 0.5:
    print("Data is not normally distributed")
else:
    print("Data is normally distributed")
```

```python
#Gerner QQ plot for sample mean 
from scipy import stats


stats.probplot(df['Glucose'], dist = 'norm', plot = plt)
plt.title('QQ plot for population Mean')
plt.show()
```

```python
#Central limit theorm 
sample_mean = []
n_sample = 100
sample_size = 30

for _ in range(n_sample):
    
    sample = df['Glucose'].sample(n=sample_size,replace=True)
    sample_mean.append(sample.mean())

print(sample_mean)
```

```python
#Gerner QQ plot for sample mean 
stats.probplot(sample_mean, dist = 'norm', plot = plt)
plt.title('QQ plot for sample Mean')
plt.xlabel('sample mean')
plt.ylabel('Frequency')
plt.show()
```

```python
shario_test = shapiro(sample_mean)
print("Shapiro-Wik p-vale: ", shario_test.pvalue)

if shario_test.pvalue < 0.5:
    print("Data is not normally distributed")
else:
    print("Data is normally distributed")
```
