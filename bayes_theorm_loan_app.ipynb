```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DataSets/loan_approval_dataset.csv")
df.head()
```

```python
N = len(df)
A = len(df[df['loan_status']=='Approved'])
B = len(df[df['cibil_score'] > 700])
```

```python
#A and B approved and good cibil score
A_and_B = len(df[(df['loan_status']=='Approved') & (df['cibil_score'] > 700)])
```

```python
P_A = A/N
P_B = B/N
P_B_given_A = A_and_B/A

#Bayes Theorem
P_A_given_B = (P_B_given_A*P_A)/P_B

print("Total number of applicants:", N)
print("Number of approved loans (A):", A)
print("Number of applicants with good cibil score (B):", B)
print("Number of approved loans with good cibil score (A and B):", A_and_B)
print("P(A) - Probability of loan approval:", P_A)
print("P(B) - Probability of good cibil score:", P_B)
print("P(B|A) - Probability of good cibil score given loan approval:", P_B_given_A)
print("P(A|B) - Probability of loan approval given good cibil score:", P_A_given_B)
```
