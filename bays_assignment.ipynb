Assignment: For the nobel dataset, find

1. Probability of winning a Nobel Prize in Physics given the laureate is female

2. Probability of winning a Nobel Prize in Literature given the laureate was born in the USA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DataSets/nobel.csv')

print(df.columns)
df.head()
```

```python

prob_physics_given_female = len(df[(df['category'] == 'Physics') & (df['gender'] == 'female')]) / len(df[df['gender'] == 'female'])
print("Probability of Nobel Prize in Physics given Female:", prob_physics_given_female)
```
