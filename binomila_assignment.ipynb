```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1 = pd.read_csv('DataSets/ODI_20.csv')
file2 = pd.read_csv('DataSets/ODI_21.csv')
df = pd.concat([file1, file2], ignore_index=True)
df.head()
```

```python
total_matches = len(df[df['Match'].str.contains("India") & ((df['Result'] == 'Won') | (df['Result'] == 'Lost')  | (df['Result'] == 'Tied'))])
total_matches_won = len(df[ (df['Country'] == 'India') & (df['Result'] == 'Won')])
print("total_matches_won:", total_matches_won)
print(f"Total matches played by India: {total_matches}")
win_probability = total_matches_won / total_matches
print(f"Probability of India winning a match: {win_probability:.4f}")
```
