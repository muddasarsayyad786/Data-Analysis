```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
df = pd.read_csv('DataSets/salary.csv')

x = df[['YearsExperience']]
y = df['Salary']

df.head()
```

```python
#spliting data into training and testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
#fit simple Lianer Regression to train data
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Step 4: Make predicion
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#step 5: Visualize traing set results
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.xlabel('Year of Exprience')
plt.ylabel("Salary")
plt.show()
```

```python
#step 5: Visualize traing set results
import matplotlib.pyplot as plt
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, model.predict(X_test), color='blue')
plt.xlabel('Year of Exprience')
plt.ylabel("Salary")
plt.show()
```

```python
#Step 7: Make new predicition
new_X_values_df = pd.DataFrame({'YearsExperience':[1,3,5,8,12,15,20,25]})
new_y_pred = model.predict(new_X_values_df)
print(new_y_pred)
```

```python
#step 8: model Eveluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mae, mse, r2)
```
