```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DataSets/tips.csv")
```

```python
plt.figure(figsize=(12,5))
#subplot1: avg tip by day
avg_tip_day = df.groupby('day')['tip'].mean()
plt.subplot(1,2,1)
plt.bar(avg_tip_day.index, avg_tip_day.values, color=['lightblue','lightgreen','pink','orange'])
plt.title("Avarage tip by day")
plt.ylabel('Avarage tip')
# plt.show()

#subplot2: smokers vs non smoker
smoker_count=df['smoker'].value_counts()
plt.subplot(1,2,2)
plt.pie(smoker_count, labels = smoker_count.index, autopct='%1.1f%%', colors=['gold', 'lightcoral'])
plt.title('Smoker vs Non smoker')
plt.tight_layout()
plt.show()
```
