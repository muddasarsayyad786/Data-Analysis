Go through the titanic dataset and either use Python or Excel to find the following:

1. Probability that a passenger dies (Marginal probability)

2. Probability that a passenger was male and survived (Joint probability of independent events)

3. Probability that a passenger survived given that she was a female (Conditional probability or Joint probability of dependent events)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Datasets/titanic.csv')
df.head()
```

```python
survived_count = len(df[df['survived']==0])
total_count = len(df)
print("probability of survived: ",np.round((survived_count/total_count)*100),2,"%")
```

```python
male_survived = len(df[(df['gender']== 'male') & (df['survived']==1)])
print(male_survived)
print("probability of male_survived: ",np.round((male_survived/total_count)*100),2,"%")
```

```python
female_survived = len(df[(df['gender']== 'female') & (df['survived']==1)])
female_probablity = len(df[df['gender']=='female'])/len(df)
female_survived_probablity = female_survived/len(df[df['gender']=='female'])
print("probability of female_survived: ",np.round((female_survived_probablity*100),2),'%')
```
