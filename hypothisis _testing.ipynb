Consider HO = doing PG-DBDA course does not improve the chances of placements and they remain at 40%.

Werrun 4 tests - Consider x bar = (1) 80, (2) 41, (3) 45, (4) -35. consider n = 100, s = 15.

Calculate z-score for two-tailed and one-tailed tests (want to see improvement only). Compare with critical value and show results. Also show p-value comparison.



```python
import math
from scipy.stats import norm
```

```python
#given
mu = 40
sample_mean = [80, 41, 45, -35]
n=100
s=15
alpha = 0.05

#Critical values
# norm.cdf() #give brobalility
# norm.ppf() #gives z-score

z_crit_two_tailed = norm.ppf(1-alpha/2) 
z_crit_one_tailed = norm.ppf(1-alpha)

print("Critical Values: ")
print(f"Two-Tailed (alpha = 0.05): {z_crit_two_tailed:.3f}")
print(f"Two-Tailed (alpha = 0.05): {z_crit_one_tailed:.3f}")
```

```python
#functio to compute Z

def compute_z(xbar):
    return (xbar - mu)/(s/math.sqrt(n))

for xbar in sample_mean:
    # p-values
    #norm.cdf(z) gives the probability to the LEFT of z
    #So 1-norm.cdf(z) gives the probability to the RIGHT of z
    #For two-tailed: compute to the right and double
    #e.g. Z = 2, Right-tail area = 0.0228
    # So, total area = 2 x 0.0228 = 0.0456
    # For one-tailed: one-sided only
    
    print(f"===============sample mean x bar = {xbar}=========")

    z = compute_z(xbar)
    
    #For one tailed: one-side only
    p_two_tailed = 2 *(1-norm.cdf(abs(z)))
    p_one_tailed = 1 - norm.cdf(z) #improvement only (right-tail)

    print(f"Z-score: {z:.3f}")

    #Two-tailed decision
    if abs(z) > z_crit_two_tailed:
        two_tail_decision = "Reject H0 (Significant difference)"
    else:
        two_tail_decision  = "Fail to Reject H0"

    #One-tailed decision

    if z > z_crit_one_tailed:
        one_tail_decision = "Rject H0"
    else:
        one_tail_decision = "Fail to Reject H0"

    print(f"Two-tailed test:")
    print(f"p-value = {p_two_tailed:.4f}")
    print(f"p-Desion : {one_tail_decision}")
```
