```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
from scipy.stats import shapiro
```

```python
df = pd.read_csv('DataSets/tips.csv')
df.head()
```

```python
total_bill_series = df['total_bill']

#Noramlity test using Shapiro-Wilk Test
shapiro_test = shapiro(total_bill_series)
print("Shapiro-Wilk Test Statistic:", shapiro_test.statistic)
print("Shapiro-Wilk Test P-value:", shapiro_test.pvalue)
```

```python
alph = 0.05 #Common significance level
if shapiro_test.pvalue > alph:
    print("Data is normally distributed")
else:
    print("Data is not normally distributed")
```

```python
#QQ plot
from scipy import stats


stats.probplot(total_bill_series, dist ="norm", plot = plt)
plt.title("QQ plot of Total Bill")
plt.xlabel("Theoretical Quantiles(Normal Distribution)")
plt.ylabel("Sample Quantiles(Total Bill)")
plt.grid()
plt.show()
```
