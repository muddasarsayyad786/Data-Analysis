```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
commuter_time = [16, 8, 35, 24, 12, 5, 23, 14, 10, 20, 25, 30, 18, 15, 45]
data_array = np.array(commuter_time)

q1 = np.percentile(data_array,25) # 25th percentile is q1
q3 = np.percentile(data_array,75) # 75th percentile is q3

iqr = q3 -q1

lower_bond = q1-(1.5*iqr)
upper_bond = q3+(1.5*iqr)

out_liers = data_array[(data_array<lower_bond) | (data_array > upper_bond)]

print("Outliers in the data set are:", out_liers)
print("Lower Bond is:", lower_bond)
print("Upper Bond is:", upper_bond) 
print("IQR is:", iqr)
print("Q1 is:", q1)
print("Q3 is:", q3)

```

```python
#Create Box plot
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(data_array, orient= 'h')
plt.title("Box plot of commuter Timers")
plt.xlabel("Minuts")
plt.show()
```
