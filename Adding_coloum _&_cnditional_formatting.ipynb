```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("tips.csv")
df.info
```

```python
print(df.tail(3))
```

```python
#adding new colum
df["tip_percentage"] = np.round(100 * df['tip']/df['total_bill'],2)
df.head()
```

```python
#drop a coloum
df = df.drop("tip_percentage", axis = 1)#axis =1 means drop a column
#df.drop("tip_percentage", axis =1 , inplace = True) # another syntax
df.head()
```

```python
#set a coloumn as index colum
df = df.set_index("Payment ID")
df.head()

#reset 
df = df.reset_index()
df.head()
```

```python
#conditional filtering
#print(df['total_bill']<30) #gives boolean values

print(df[df['total_bill']<30])


print(df[(df['gender'] == "Male") & (df['total_bill']<30) ])
```

```python

print(df[(df['gender'] == "Male") & (df['total_bill']<30) ])
```

```python
#Sql like syntax
filtered = df.query("gender=='Male' and tip > 4")
filtered.head
```
