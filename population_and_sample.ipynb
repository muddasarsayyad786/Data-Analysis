Probability Based Sampling


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Read each CSV file into a separate DataFrame
df_2015 = pd.read_csv('DataSets/marathon_results_2015.csv')
df_2016 = pd.read_csv('DataSets/marathon_results_2016.csv')
df_2017 = pd.read_csv('DataSets/marathon_results_2017.csv')

# 2. Concatenate the DataFrames by passing them in a list
df = pd.concat([df_2015, df_2016, df_2017], ignore_index=True)

# 3. Display the head of the combined DataFrame
print(df.head())
df.shape
```

```python
#simple sampling method (select 10% of record randomly)
sample_df = df.sample(frac=0.1, random_state=42) #frac 0.1 for getting 10% data and randon_state to keep random data fix
print("Sample data shape: ",sample_df.shape)
print("Population Data shape: ",df.shape)
```

```python
#Systematic Ramdom Sampling Method
sample_df1 = df.iloc[::7, :] #df.iloc[row_indexer(start:end:step), column_indexer(start:end:step)]
print("Sample data shape: ",sample_df1.shape)
print("Population Data shape: ",df.shape)
```

```python
#Strtified Random sampling
total_male = len(df[df['M/F']=='M'])
total_female = len(df[df['M/F']=='F'])

sample_size = int(0.1*df.shape[0])
sample_male = int(0.1*total_male)
sample_female = int(0.1*total_female)

print(sample_size,sample_male, sample_female)

#genrate proportion samples of male and female
male_sample = df[df['M/F']=='M'].sample(n=sample_male, random_state=42)
female_sample = df[df['M/F']=='F'].sample(n=sample_female,random_state=42)
print("Lenths of male and female: ",len(male_sample), len(female_sample))


#combine male and female sample in single frame

sample_df2 = pd.concat([male_sample,female_sample])
sample_row_count = len(sample_df2)
print(sample_df2.head())
print("length of sample: ",sample_row_count)



```

```python
#cluster Smapling : find unqiue countries the athelets are from

unique_countries = df['Country'].unique()
num_coutry = len(unique_countries)
print(unique_countries,num_coutry)

#randomly select half othe country
np.random.seed(42)
selected_country = np.random.choice(unique_countries, size=num_coutry//2, replace=False)
#choice select random values , replce false dont allow repeact of values
print("selected countries: ", selected_country)

#create new df that contains record of selected countries only
sample_df3 = df[df['Country'].isin(selected_country)]
sample_row_count = len(sample_df3)
print("length od sample df: ",sample_row_count)
sample_df3.head
```
