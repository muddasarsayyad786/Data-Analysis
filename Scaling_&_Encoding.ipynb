```python
#Scaling example
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

```

```python
df = pd.read_csv('DataSets/tips.csv')
total_bill = df[['total_bill']]
df.head()
```

```python
#1.Stander Scaling
stander_scaling = StandardScaler()
df['total_bill_stander_scaled'] = stander_scaling.fit_transform(total_bill)
df.head()
```

```python
#2. Min-max Scaling
minmax_scaler = MinMaxScaler()
df['total_bill_minmax_scaled'] = minmax_scaler.fit_transform(total_bill)
df.head()
```

```python
pd.set_option('display.max_columns', None)
print(df[['total_bill', 'total_bill_stander_scaled', 'total_bill_minmax_scaled']].head(10))
pd.reset_option('display.max_columns')
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
df = pd.read_csv("DataSets/tips.csv")
print("Original 'day' coloumn:\n ", df['day'].head())
```

```python
#1 one hot encoding
ohe = OneHotEncoder(sparse_output=False, drop=None)
ohe_array = ohe.fit_transform(df[['day']])
```

```python
#oloumn names created by OneHotEncoder
ohe_cols = ohe.get_feature_names_out(['day'])
#Create new dataframe woth ohe coloumns
df_ohe = df.join(pd.DataFrame(ohe_array, columns = ohe_cols, index=df.index))
```

```python
print("One hot encoding :", df_ohe[ohe_cols].head())
print("One hot Encoding: ", df_ohe[ohe_cols].tail())
```

```python
ohe_drop = OneHotEncoder(sparse_output=False, drop='first')
ohe_array2 = ohe_drop.fit_transform(df[['day']])
ohe_cols2 = ohe_drop.get_feature_names_out(['day'])
df_ohe2 = df.join(pd.DataFrame(ohe_array2, columns = ohe_cols2, index=df.index))
print(df_ohe2[ohe_cols2].head())
print(df_ohe2[ohe_cols2].tail())
```

```python
#2 lable encoding
label_encoder = LabelEncoder()
df['day_label'] = label_encoder.fit_transform(df['day'])
print(df_ohe2[ohe_cols2].head())
print(df_ohe2[ohe_cols2].tail())
```
