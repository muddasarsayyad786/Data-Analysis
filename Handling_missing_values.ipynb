```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("DataSets/titanic.csv")
df1.head
```

```python
#Manual Methods: use pandas
df1['age_mean']=df1['age'].fillna(df1['age'].mean())
df1['age_mode']=df1['age'].fillna(df1['age'].mode()[0])
df1['age_median']=df1['age'].fillna(df1['age'].mean())
df1['age_constant']=df1['age'].fillna(0)
df1.head
```

```python


#handling nulls using SimpleImputer of sklearn library
from sklearn.impute import SimpleImputer

# create imputers for different strategies
imp_mean = SimpleImputer(strategy='mean')
imp_mode = SimpleImputer(strategy='most_frequent')
imp_median = SimpleImputer(strategy='median')
imp_constant = SimpleImputer(strategy='constant', fill_value=0)

# fit_transform expects 2D input; use df1[['age']] to keep it 2D
df1['imputer_mean'] = imp_mean.fit_transform(df1[['age']])
df1['imputer_mode'] = imp_mode.fit_transform(df1[['age']])
df1['imputer_median'] = imp_median.fit_transform(df1[['age']])
df1['imputer_constant'] = imp_constant.fit_transform(df1[['age']])
df1.head
```
