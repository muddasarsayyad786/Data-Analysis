```python
import pandas as pd
from scipy import stats
```

```python
#2
df1 = pd.read_csv('DataSets/marathon_results_2015.csv')
df2 = pd.read_csv('DataSets/marathon_results_2016.csv')
df3  = pd.read_csv('DataSets/marathon_results_2017.csv')

df = pd.concat([df1, df2, df3], ignore_index=True)
df.head()
```

```python
#H0 = Younger age and older age runners have similar  no  difference in avg space

df[['Hrs', 'Mins', 'Sec']] = df['Pace'].str.split(':', expand=True)

# normalize & convert to integers safely
for col in ['Hrs', 'Mins', 'Sec']:
    df[col] = pd.to_numeric(df[col].astype(str).str.strip(), errors='coerce').fillna(0).astype(int)

# compute seconds
df['Pace_in_sec'] = df['Hrs']*3600 + df['Mins']*60 + df['Sec']

younger_age = df[df['Age']<=30]['Pace_in_sec'].dropna().sample(n=50, replace=False)
older_age = df[df['Age']>30]['Pace_in_sec'].dropna().sample(n=50, replace=False)

t_statistic, p_value = stats.ttest_ind(younger_age, older_age)
alpha = 0.05
print("p_value: ", p_value)

if p_value < alpha:
    print("Reject H0")
else:
    print("Do not Reject H0")
```

```python
#3
#H0: parantal education not effects on childs math marks
df = pd.read_csv('DataSets/StudentsPerformance.csv')
df.head()
```

```python
print(df['parental level of education'].unique())
```

```python
bachelor_degree = df[df['parental level of education'] == "bachelor's degree"]['math score']
some_college = df[df['parental level of education'] == "some college"]['math score']
master_degree = df[df['parental level of education'] == "master's degree"]['math score']
associate_degree = df[df['parental level of education'] == "associate's degree"]['math score']
high_school = df[df['parental level of education'] == "high school"]['math score']
some_high_school = df[df['parental level of education'] == "some high school"]['math score']

fstat, p_value = stats.f_oneway(bachelor_degree, some_college, master_degree, associate_degree, high_school, some_high_school)
print("P-value: ",p_value)

alpha = 0.05
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```
