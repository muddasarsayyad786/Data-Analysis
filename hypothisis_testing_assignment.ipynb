1. A courier company delivers parcels in a particular area in 30 minutes. A new GPS system installed has reduced the average time to 28 minutes, when 100 trucks were studied, with a sample standard deviation of 4.2 minutes. Run a two-tailed test and a one-tailed (left-tailed) test.

2. A battery manufacturer claims that a battery which used run for 30 hours now runs for more tha 30 hours due to new technology. 200 batteries were studied. Average battery time was found to be 29.5 hours with a standard deviation of 8 hours. Run a one-tailed (right-tailed) test.

```python
import scipy.stats as stats
import math

# Given data
mu_0 = 30              # population mean (old mean)
x_bar = 28             # sample mean
s = 4.2                # sample standard deviation
n = 100                # sample size

# Standard error
se = s / math.sqrt(n)

# Test statistic (z-score)
z = (x_bar - mu_0) / se

print("Z-statistic:", z)

# TWO-TAILED TEST (H1: μ ≠ 30)
p_two_tailed = 2 * stats.norm.cdf(abs(z))
print("Two-tailed p-value:", p_two_tailed)

# LEFT-TAILED TEST (H1: μ < 30)
p_left_tailed = stats.norm.cdf(z)
print("Left-tailed p-value:", p_left_tailed)
```
