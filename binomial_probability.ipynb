#Simulation of binomial distribution on titanic dataset - If we picked 100 people randomly from this Titanic data, what is the probability distribution of the number of survivors? Imagin we are studying the Titanic disaster and want to understand the likelihood of survival. We look at real passenger data and calculate the overall survival rate that a randomly chosen person survived. Now, suppose we randomly select 100 passengers - what are the chances that 50 of them survived? Or 60? Or 70? That is exa that is, the probability what the binomial distribution helps us explore. It shows us how many survivors we might expect out of 100 random passengers, and how likely each outcome is, based on what actually happened in the real data

```python
#simulation binomial distribuation on titanic data set
df = pd.read_csv('DataSets/titanic.csv')
df.head()
```

```python
df['survived_binary'] = df['survived'].apply(lambda x: 1 if x==1 else 0)
n= len(df)
k=df['survived_binary']
p = k / n
print(n, k, p)
```

```python
#simulate 100 hypothesis passenger
trials = 100
x = np.arange(0, trials + 1)

#pmf and cdf
pmf_values = binom.pmf(x, trials, p)
cdf_values = binom.cdf(x,trials,p)

plt.figure(figsize=(12,6))
#PMF plot
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, color='skyblue', alpha=0.7)
plt.title('Binomial PMF - Titanic Survival')
```
